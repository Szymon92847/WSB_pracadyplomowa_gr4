Import bibliotek

```python
import pandas as pd
from sqlalchemy import create_engine
import seaborn as sb
import matplotlib as plt
import pyarrow as pa
```

Import danych z bazy danych

```python
connection_string = (
    "mssql+pyodbc://localhost/DiamondsDB"
    "?driver=ODBC+Driver+17+for+SQL+Server"
    "&trusted_connection=yes"
)
engine = create_engine(connection_string)
query = "SELECT * FROM DIAMONDS_CLEANED"
df_diamonds = pd.read_sql_query(query, engine)
df_diamonds.head(15)
```

Sprawdzenie typów danych i pustych wierszy. Usuniecie pustych wierszy.

```python
df_diamonds.info()
```

```python
# Sprawdzenie, czy są wartości null
print("Czy dane zawierają wartości null?")
print(df_diamonds.isnull().sum())

# Usunięcie wierszy zawierających wartości null
df_diamonds = df_diamonds.dropna()
```

Zmiana nazw kolumn

```python
df_diamonds.columns = ['Index', 'Carat','Cut', 'Color', 'Clarity', 'Depth', 'Table', 'Price', 'x', 'y', 'z']
df_diamonds.head(15)
```

```python
df_diamonds["Size"] = df_diamonds["x"]*df_diamonds["y"]*df_diamonds["z"]
df_diamonds["Size"] = df_diamonds["Size"].round(2)
df_diamonds.head(15)
```

```python
cut_mapping = {
    "Ideal": 5,
    "Premium": 4,
    "very Good": 3,
    "Good": 2,
    "Fair": 1
}
df_diamonds["Cut_number_classification"] = df_diamonds["Cut"].map(cut_mapping)
df_diamonds.head(15)
```

```python
color_mapping = {
    "D": 7,
    "E": 6,
    "F": 5,
    "G": 4,
    "H": 3,
    "I": 2,
    "J": 1
}
df_diamonds["Color_number_classification"] = df_diamonds["Color"].map(color_mapping)
df_diamonds.head(15)
```

```python
clarity_mapping = {
    "IF": 8,
    "VVS1": 7,
    "VVS2": 6,
    "VS1": 5,
    "VS2": 4,
    "SI1": 3,
    "SI2": 2,
    "I1": 1
}
df_diamonds["Clarity_number_classification"] = df_diamonds["Clarity"].map(clarity_mapping)
df_diamonds.head(15)
```

Segmentacja na podstawie wagi/caratu diamentu. Analiza ABC

```python
carat_segmentation = df_diamonds[["Index", "Carat"]]
carat_segmentation = carat_segmentation.sort_values(by='Carat', ascending=False)
total_carat = carat_segmentation['Carat'].sum()

carat_segmentation['Percentage_share_carat'] = (carat_segmentation['Carat'] / total_carat) * 100
carat_segmentation['Cumulative_percentage_share_carat'] = carat_segmentation['Percentage_share_carat'].cumsum()

def abc_carat_segmentation(value):
    if value <= 80:
        return 'A'
    elif value <= 95:
        return 'B'
    else:
        return 'C'

carat_segmentation['Carat_ABC_segmentation'] = carat_segmentation['Cumulative_percentage_share_carat'].apply(abc_carat_segmentation)
carat_segmentation.drop(columns=["Carat","Percentage_share_carat","Cumulative_percentage_share_carat"], inplace=True)
carat_segmentation.head(60000)
```

Segmentacja na podstawie cech jakościowych(cut, color, clarity) diamentu. Analiza ABC
